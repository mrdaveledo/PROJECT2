### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
file_nested_list.choice_description = file_nested_list.choice_description.str.replace("[", "")
file_nested_list.choice_description = file_nested_list.choice_description.str.replace("]", "")
file_nested_list.head()
```

```python
burrito = file_nested_list[(file_nested_list.item_name == 'Chicken Burrito') | (file_nested_list.item_name == 'Veggie Burrito') | (file_nested_list.item_name == 'Steak Burrito')]

burrito.head()

#burrito.index
burrito_orders = burrito.shape[0]
print (burrito_orders)

split_columns = burrito.choice_description.str.split(',', expand=True)

split_columns.columns = ["col0", "col1", "col2", "col3", "col4", "col5", "col6", "col7", "col8", "col9"]

col0 = split_columns.col0.count()
col1 = split_columns.col1.count()
col2 = split_columns.col2.count()
col3 = split_columns.col3.count()
col4 = split_columns.col4.count()
col5 = split_columns.col5.count()
col6 = split_columns.col6.count()
col7 = split_columns.col7.count()
col8 = split_columns.col8.count()
col9 = split_columns.col9.count()

toppings_total = col0+col1+col2+col3+col4+col5+col6+col7+col8+col9

print (toppings_total)

averagetoppings = round((toppings_total / burrito_orders), 2)

print (averagetoppings)
```
